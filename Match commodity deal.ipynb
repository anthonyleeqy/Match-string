```python
def get_closest_match(x, list_strings):
    best_match = None
    highest_jw = 0
    for current_string in list_strings:
        current_score = jellyfish.jaro_winkler(x, current_string)

        if(current_score > highest_jw):
            highest_jw = current_score
            best_match = current_string

    return best_match
```

```python
GSThisMonth['Secdb Id'] = GSThisMonth['Secdb Id'].astype(str)
GSLastMonth['Secdb Id'] = GSLastMonth['Secdb Id'].astype(str)
```

```python
GSThisMonth['Match'] = GSThisMonth['Secdb Id'].map(lambda x: get_closest_match(x, GSLastMonth['Secdb Id']))
```

```python
Result = GSLastMonth.merge(GSThisMonth,left_on='Secdb Id', right_on='Match', how='right')
Result.head()
```

```python
Result.to_csv("ResultCommodity.csv")
```
